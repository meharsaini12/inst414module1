```python
import requests
import json 
import pandas as pd
import matplotlib.pyplot as plt
from openpyxl.workbook import Workbook
import statistics
```

```python
#Open and request API through URL

url = "http://makeup-api.herokuapp.com/api/v1/products.json?product_type=blush"
makeup_products = requests.get(url)
```

```python
#Convert API into json file
makeup_json = json.loads(makeup_products.text)

#Change json file into dataframe
makeup_df = pd.json_normalize(makeup_json)
makeup_df = makeup_df.fillna(0)
makeup_df
```

```python
#Average maybelline blush price

maybelline_df = makeup_df.loc[makeup_df['brand']=='maybelline']
avg_df = maybelline_df[['brand', 'price']]

price_list = []
for index, row in avg_df.iterrows():
    price_list.append(row['price'])

def avg(lists):
    up_list = []
    for i in lists:
        up_list.append(float(i))
    return statistics.fmean(up_list)

blush_may = avg(price_list)
#print(blush_may)
```

```python
#Average nyx blush price

nyx_df = makeup_df.loc[makeup_df['brand']=='nyx']
nyxavg_df = nyx_df[['brand', 'price']]


price_list = []
for index, row in nyxavg_df.iterrows():
    price_list.append(row['price'])

def avg(lists):
    up_list = []
    for i in lists:
        up_list.append(float(i))
    return statistics.fmean(up_list)

blush_nyx = avg(price_list)
#print(blush_nyx)
```

```python
#Average clinique blush price

clinique_df = makeup_df.loc[makeup_df['brand']=='clinique']
cliniqueavg_df = clinique_df[['brand', 'price']]

price_list = []
for index, row in cliniqueavg_df.iterrows():
    price_list.append(row['price'])

def avg(lists):
    up_list = []
    for i in lists:
        up_list.append(float(i))
    return statistics.fmean(up_list)

blush_clinique = avg(price_list)
#print(blush_clinique)
```

```python
#Build data frame of average blush prices
  
average_data = [['maybelline', blush_may ], ['nyx', blush_nyx ], ['covergirl', blush_covergirl], 
               ['physicians formula', blush_pf], ['clinique', blush_clinique]]
  
blushavg_df = pd.DataFrame(average_data, columns=['brand', 'average price'])
#blushavg_df

#Create bar graph visualization
bar = blushavg_df.plot.bar(x='brand', y='average price', rot=0, xlabel = 'Makeup Brand', 
                           ylabel = 'Average Price', 
                           title = 'Average Price of Blushes based on Brand')
```
